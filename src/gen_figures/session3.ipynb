```python
import sys
from pathlib import Path

# Add src directory to path for imports
sys.path.insert(0, str(Path().resolve().parent))

import numpy as np
import matplotlib.pyplot as plt
from simulators import simulate

# Set style for minimal plots
plt.rcParams.update({
    'figure.facecolor': 'None',
    'axes.facecolor': 'None',
})

# Fixed parameters for the Lotka-Volterra model (observation)
alpha = 0.1
beta = 0.02
delta = 0.01
gamma = 0.1
parameters_obs = np.array([alpha, beta, delta, gamma])

# Simulate the observation
time = np.arange(0, 200, 0.1)
populations_obs = simulate(parameters_obs)
prey_obs = populations_obs[:, 0]
predator_obs = populations_obs[:, 1]

# Simulate two gray traces (slightly different parameters)
np.random.seed(123)
params_gray1 = parameters_obs * (1 + 0.25 * np.random.randn(4))
params_gray2 = parameters_obs * (1 + 0.20 * np.random.randn(4))
pop_gray1 = simulate(params_gray1)
pop_gray2 = simulate(params_gray2)

# Simulate traces close to observation (violet) - small perturbations
np.random.seed(456)
params_violet1 = parameters_obs * (1 + 0.03 * np.random.randn(4))
params_violet2 = parameters_obs * (1 - 0.02 * np.random.randn(4))
pop_violet1 = simulate(params_violet1)
pop_violet2 = simulate(params_violet2)

# Simulate trace far from observation (magenta) - very different
params_magenta = parameters_obs * np.array([4.0, 0.3, 3.0, 0.3])
pop_magenta = simulate(params_magenta)

# --- Plot 1: Observation trace (green) - predator over time ---
fig, ax = plt.subplots(figsize=(3, 2.5))
ax.plot(time, predator_obs, color='#228B22', linewidth=3, alpha=0.9)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
plt.tight_layout()
# plt.savefig('trace_observation.png', dpi=150, transparent=True, bbox_inches='tight', pad_inches=0.05)
plt.show()

# --- Plot 2: Two simulation traces (gray tones) - predator over time ---
fig, ax = plt.subplots(figsize=(3, 2.5))
ax.plot(time, pop_gray1[:, 1], color='#404040', linewidth=3, alpha=0.9)
ax.plot(time, pop_gray2[:, 1], color='#606060', linewidth=3, alpha=0.9)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
plt.tight_layout()
# plt.savefig('trace_simulations.png', dpi=150, transparent=True, bbox_inches='tight', pad_inches=0.05)
plt.show()

# --- Plot 3a: Violet trace 1 (close to observation) - predator over time ---
fig, ax = plt.subplots(figsize=(3, 2.5))
ax.plot(time, pop_violet1[:, 1], color='#6A0DAD', linewidth=3, alpha=0.9)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
plt.tight_layout()
# plt.savefig('trace_violet1.png', dpi=150, transparent=True, bbox_inches='tight', pad_inches=0.05)
plt.show()

# --- Plot 3b: Violet trace 2 (close to observation) - predator over time ---
fig, ax = plt.subplots(figsize=(3, 2.5))
ax.plot(time, pop_violet2[:, 1], color='#5B3A8C', linewidth=3, alpha=0.9)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
plt.tight_layout()
# plt.savefig('trace_violet2.png', dpi=150, transparent=True, bbox_inches='tight', pad_inches=0.05)
plt.show()

# --- Plot 3c: Magenta trace (far from observation) - predator over time ---
fig, ax = plt.subplots(figsize=(3, 2.5))
ax.plot(time, pop_magenta[:, 1], color='#C71585', linewidth=3, alpha=0.9)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
plt.tight_layout()
# plt.savefig('trace_magenta.png', dpi=150, transparent=True, bbox_inches='tight', pad_inches=0.05)
plt.show()
```
